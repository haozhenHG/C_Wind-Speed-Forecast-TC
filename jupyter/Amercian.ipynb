````python
temp_cols = ['temp1', 'temp2', 'temp3']
datac['temp_gradient'] = datac[temp_cols].diff(axis=1).mean(axis=1)  # æ—¶é—´æ¢¯åº¦
````

````python
# æž„å»ºçƒ­åŠ›å­¦ç‰¹å¾
datac['heat_index'] = 0.5 * (datac['temp2'] + 0.1 * datac['temp2'] * datac['temp1'])
````

````python
# è´Ÿè½½ç›¸å…³
# æ»žåŽç‰¹å¾
datac['load_lag24'] = datac['load'].shift(24)  # æ˜¨æ—¥åŒæœŸè´Ÿè½½

# æ»šåŠ¨ç»Ÿè®¡é‡
# è®¡ç®—è¿‡åŽ» 1 åˆ° 7 å¤©çš„æ»šåŠ¨  å‡å€¼ æ ‡å‡†å·®
for days in range(1, 8):
    window_size = 24 * days
    datac[f'load_roll_mean_{days}d'] = datac['load'].rolling(window_size).mean()    
    datac[f'load_roll_std_{days}d'] = datac['load'].rolling(window_size).std()
````

````python
import pandas as pd
from sklearn.model_selection import train_test_split
import lightgbm as lgb

def get_feature_importances(data, shuffle, seed=None):
    # å‰”é™¤ Unnamed: 0 åˆ—
    features = [i for i in data.columns if i not in ['load']]
    y = data['load'].copy()
    x = data[features]

    if shuffle:
        # Here you could as well use a binomial distribution
        y = data['load'].copy().sample(frac=1.0)

    xtrain, xval, ytrain, yval = train_test_split(x, y, random_state=42, test_size=0.25)
    model = lgb.LGBMRegressor(
        num_leaves=32,
        max_depth=7,
        learning_rate=0.08,
        n_estimators=1000,
        importance_type='split',
        subsample=0.8,
        feature_fraction=0.8,
        reg_alpha=0.5,
        reg_lambda=0.5,
        random_state=2021,
        objective='regression',
        metric='mse',
        device='cpu'  # ä¿®æ”¹ä¸ºä½¿ç”¨ CPU è¿›è¡Œè®¡ç®—
    )

    # Fit the model
    model = model.fit(
        xtrain,
        ytrain,
        eval_metric="mse",
        eval_set=[(xval, yval)]
    )

    # Get feature importances
    imp_df = pd.DataFrame()
    imp_df["feature"] = list(features)
    imp_df["importance_split"] = model.feature_importances_

    return imp_df

# è°ƒç”¨å‡½æ•°èŽ·å–ç‰¹å¾é‡è¦æ€§
result = get_feature_importances(datacc, shuffle=False)
````

````python
# æŒ‰ç…§é‡è¦æ€§é™åºæŽ’åº
result = result.sort_values(by='importance_split', ascending=False).reset_index(drop=True)

print('ç‰¹å¾é‡è¦æ€§ï¼š')
result
````
