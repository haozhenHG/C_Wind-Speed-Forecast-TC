## 二、深度特征工程

````python
datac.head()
````

````python
# 周期性编码
datac['week_sin'] = np.sin(2*np.pi*datac['weekday']/7)
datac['week_cos'] = np.cos(2*np.pi*datac['weekday']/7)

# 时间上下文特征
datac['hour'] = datac.index.hour
datac['is_working_hour'] = ((datac['hour']>=8) & (datac['hour']<=18)).astype(int)
````

````python
datac.head(24)
````

````python
datac[['temp1','temp2',	'temp3']].describe()
````

````python
# 绘制折线图
plt.figure(figsize=(13, 3))
for col in ['temp1', 'temp2', 'temp3']:
    plt.plot(datac.index, datac[col], label=col)

# 添加标题和标签
plt.title('Temperature Variation Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
# plt.grid(True)

# 显示图形
plt.show()
````

````python
temp_cols = ['temp1', 'temp2', 'temp3']
datac['temp_gradient'] = datac[temp_cols].diff(axis=1).mean(axis=1)  # 时间梯度
````

````python
# 构建热力学特征
datac['heat_index'] = 0.5 * (datac['temp2'] + 0.1 * datac['temp2'] * datac['temp1'])
````

````python
# 负载相关
# 滞后特征
datac['load_lag24'] = datac['load'].shift(24)  # 昨日同期负载

# 滚动统计量
# 计算过去 1 到 7 天的滚动  均值 标准差
for days in range(1, 8):
    window_size = 24 * days
    datac[f'load_roll_mean_{days}d'] = datac['load'].rolling(window_size).mean()    
    datac[f'load_roll_std_{days}d'] = datac['load'].rolling(window_size).std()
````

````python
datac.to_csv('American_new.csv')
````

由于构造了滞后特征相关的信息    导致原始数据出现不争气的情况   -- 直接删除  原始数据集的影响小

````python
datacc = datac.copy()
````

````python
datacc.dropna(inplace=True)
datacc.head()
````

````python
import pandas as pd
from sklearn.model_selection import train_test_split
import lightgbm as lgb

def get_feature_importances(data, shuffle, seed=None):
    # 剔除 Unnamed: 0 列
    features = [i for i in data.columns if i not in ['load']]
    y = data['load'].copy()
    x = data[features]

    if shuffle:
        # Here you could as well use a binomial distribution
        y = data['load'].copy().sample(frac=1.0)

    xtrain, xval, ytrain, yval = train_test_split(x, y, random_state=42, test_size=0.25)
    model = lgb.LGBMRegressor(
        num_leaves=32,
        max_depth=7,
        learning_rate=0.08,
        n_estimators=1000,
        importance_type='split',
        subsample=0.8,
        feature_fraction=0.8,
        reg_alpha=0.5,
        reg_lambda=0.5,
        random_state=2021,
        objective='regression',
        metric='mse',
        device='cpu'  # 修改为使用 CPU 进行计算
    )

    # Fit the model
    model = model.fit(
        xtrain,
        ytrain,
        eval_metric="mse",
        eval_set=[(xval, yval)]
    )

    # Get feature importances
    imp_df = pd.DataFrame()
    imp_df["feature"] = list(features)
    imp_df["importance_split"] = model.feature_importances_

    return imp_df

# 调用函数获取特征重要性
result = get_feature_importances(datacc, shuffle=False)
````

````python
# 按照重要性降序排序
result = result.sort_values(by='importance_split', ascending=False).reset_index(drop=True)

print('特征重要性：')
result
````

````python
# 设置负号显示
plt.rcParams['axes.unicode_minus'] = False
# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
# 创建柱状图
plt.figure(figsize=(8,5))
sns.barplot(x='importance_split', y='feature', data=result)

# # 添加数据标签
# for i, v in enumerate(result['importance_split']):
#     plt.text(v + 3, i + 0.25, str(v), color='black')

# 添加标题和标签
plt.title('特征重要性可视化')
plt.xlabel('特征重要性得分')
plt.ylabel('特征名称')

# 显示图形
plt.show()
````

通过上述信息 并结合实际情况 
```
1. 核心影响特征
hour（小时）：重要性得分最高，说明一天内的不同小时对目标变量（如负载）影响显著，反映出时间周期性规律（如早晚高峰用电差异）对结果的强关联性。
负载相关衍生特征：
load_roll_mean_1d（1 天滚动均值）、load_lag24（24 小时滞后负载）、load_roll_std_1d（1 天滚动标准差）等特征重要性靠前，表明历史负载数据的均值、滞后值、波动情况对当前目标变量有较强预测能力，体现了负载数据的时序依赖性。
2. 环境与时间特征
温度相关特征：temp1、temp2、temp3及temp_gradient（温度梯度）重要性中等，说明温度及其变化趋势对目标变量存在影响，可能反映温度与负载（如空调使用导致的用电变化）的关联。
```

### stl + vmd 探索性分析

````python
df = pd.read_csv('./American_final_to_model.csv',index_col=0) # 列 0 作为时间索引戳
df.head()
````

````python
df.shape
````

````python
df['load'].plot(figsize=(16,4))
````

````python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
from vmdpy import VMD
````

````python
# 1. STL分解
stl = STL(df['load'].values, period=168, robust=True)
result = stl.fit() # 季节性趋势分解
trend = result.trend
seasonal = result.seasonal
residual = result.resid # 残差是原始时间序列数据减去趋势和季节性成分后剩余的部分，它包含了时间序列中无法由趋势和季节性解释的随机波动或噪声。
````

````python
plt.figure(figsize=(16,6))
ax = plt.subplot(3,1,1)
ax.plot(trend)
ax = plt.subplot(3,1,2)
plt.plot(seasonal)
ax = plt.subplot(3,1,3)
plt.plot(residual)
plt.tight_layout()
plt.show()
````
